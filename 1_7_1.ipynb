```python
import numpy as np

w = np.array([15/7, -9/7, -1]) # разделяющая линия определяется параметрами
x_test = np.array([(-8, -4), (-2, 2), (4, 8), (6, 3)]) # (признаки образов: x0, x1, x2)
y_test = np.array([1, 1, -1, -1]) # (метки класса)

margin = [] # <w, x> * y - функция отступа
for i in range(len(x_test)):
    margin.append((w[1] * x_test[i][0] + w[2] * x_test[i][1] + w[0]) * y_test[i])

print(margin)
```
