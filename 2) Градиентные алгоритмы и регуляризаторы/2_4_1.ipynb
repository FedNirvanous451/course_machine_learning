```python
import numpy as np


def func(x):
    return -0.5 * x + 0.2 * x ** 2 - 0.01 * x ** 3 - 0.3 * np.sin(4*x)

def df(x):
    return 0.2 * 2 * x - 0.01 * 2 * x ** 2 - 0.3 * 4 * np.cos(4*x) - 0.5

def momentum():
    x0 = -3.5 
    m0 = 0.1 # шаг сходимости
    N = 200 # число итераций
    j1 = 0.8 # параметр экспон. скольз. среднего
    x = x0
    v = 0 # начальное значение переменной накопления скольз. среднего
    for _ in range(N):
        v = j1 * v + (1 - j1) * m0 * df(x)
        x -= v
    return x

def gradient():
    x0 = -3.5 
    m0 = 0.1 # шаг сходимости
    N = 200 # число итераций
    x = x0
    for _ in range(N):
        x -= m0 * df(x)
    return x

print(momentum(), gradient())
```
