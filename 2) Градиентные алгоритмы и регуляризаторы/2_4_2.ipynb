```python
import numpy as np


def func(x):
    return 0.4 * x + 0.1 * np.sin(2*x) + 0.2 * np.cos(3*x)


def df(x):
    return 0.4 + 0.1 * 2 * np.cos(2*x) - 0.2 * 3 * np.sin(3*x) 


def momentum():
    x0 = 4
    m0 = 1.0 #  
    N = 500 #  
    j1 = 0.7 #  . . 
    x = x0
    v = 0 #     . 
    for _ in range(N):
        v = j1 * v + (1 - j1) * m0 * df(x)
        x -= v
    return x

def nesterov():
    x0 = 4
    m0 = 1.0 #  
    N = 500 #  
    j1 = 0.7 #  . . 
    x = x0
    v = 0 #     . 
    for _ in range(N):
        v = j1 * v + (1 - j1) * m0 * df(x - j1 * v)
        x -= v
    return x

def gradient():
    x0 = 4
    m0 = 1.0 #  
    N = 500 #  
    x = x0
    for _ in range(N):
        x -= m0 * df(x)
    return x

print(nesterov(), gradient(), momentum())
```
