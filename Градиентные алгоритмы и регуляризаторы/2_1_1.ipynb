```python
import numpy as np


def func(x):
    return 0.5 * x + 0.2 * x ** 2 - 0.1 * x ** 3 # заданная функция f(x)


def df(x):
    return 0.5 + 0.2 * 2 * x - 0.1 * 3 * x ** 2 # производная f(x)


coord_x = np.arange(-5.0, 5.0, 0.1) # значения по оси абсцисс
coord_y = func(coord_x) # значения по оси ординат (значения функции)


m = 0.01 # шаг сходимости
def gradient(x, N):
    # рекуррентная реализация градиентного алгоритма
    if N == 200:
        return x - m * df(x)
    if df(x) != 0:
        return gradient(x - m * df(x), N + 1)   
    return x - m * df(x)

print(gradient(-4, 0))

```

```python
# реализация кода с наглядным выводом работы алгоритма
import numpy as np
import matplotlib.pyplot as plt
import time

def func(x):
    return 0.5 * x + 0.2 * x ** 2 - 0.1 * x ** 3

def df(x):
    return 0.5 + 0.4 * x - 0.3 * x ** 2

coord_x = np.arange(-5.0, 5.0, 0.1) # значения по оси абсцисс
coord_y = func(coord_x) # значения по оси ординат (значения функции)

x0 = -4
lmb = 0.01
N = 200

plt.ion()
fig, ax = plt.subplots()
ax.grid(True)
ax.plot(coord_x, coord_y)
point = ax.scatter(x0, func(x0), c="red")

x = x0
for _ in range(N):
    x -= lmb * df(x)

    point.set_offsets([x, func(x)])
    fig.canvas.draw()
    fig.canvas.flush_events()
    time.sleep(0.01)

plt.ioff()
print(x)
ax.scatter(x, func(x), c="green")
plt.show()
```
