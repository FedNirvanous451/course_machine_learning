```python
import numpy as np


def func(x):
    return 2 * x + 0.1 * x ** 3 + 2 * np.cos(3*x)


def df(x):
    return 2 + 0.1 * 3 * x ** 2 - 2 * 3 * np.sin(3*x)


def RMSProp():
    x0 = 4
    m0 = 0.5 #  
    N = 200 #  
    a = 0.8
    x = x0
    e = 0.01
    G = 0
    for _ in range(N):
        G = a * G + (1 - a) * df(x) * df(x)
        x -= m0 * df(x) / (np.sqrt(G) + e)
    return x

print(RMSProp())
```
