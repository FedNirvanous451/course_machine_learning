```python
import numpy as np
from sklearn import svm

def func(x):
    return np.sin(0.5*x) + 0.2 * np.cos(2*x) - 0.1 * np.sin(4 * x) - 2.5


def model(w, x):
    return w[0] + w[1] * x + w[2] * x ** 2 + w[3] * x ** 3 + w[4] * np.cos(x) + w[5] * np.sin(x)


#  
coord_x = np.arange(-4.0, 6.0, 0.1)
coord_y = func(coord_x)

x_train = np.array([[x, x**2, x**3, np.cos(x), np.sin(x)] for x in coord_x])
y_train = coord_y

#   
svr = svm.SVR(kernel='linear')    # SVR   
svr.fit(x_train, y_train)

w12 = svr.coef_[0]
w0 = svr.intercept_
w = np.concatenate((w0, w12))

ar_ones = np.ones(len(coord_x))
new_x = np.hstack((ar_ones[:, np.newaxis], x_train))

Q = np.mean((w @ new_x.T - y_train)**2)
Q
```
