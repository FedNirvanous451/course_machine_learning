```python
import numpy as np

w = np.array([-16/9, -4/9, 1]) # (параметры модели: w0, w1, w2 - ориентация разделяющей линии)
x_test = np.array([[1, -5, 2], [1, -4, 6], [1, 3, 2], [1, 3, -3], [1, 5, 6], [1, 9, 2]]) # (признаки образов: x0, x1, x2)
y_test = np.array([1, 1, 1, -1, -1, -1]) # (метки класса)

# здесь продолжайте программу
margin = [] # <w, x> * y - функция отступа
for i in range(len(x_test)):
    margin.append((np.dot(x_test[i], w)) * y_test[i])

print(margin)
```
