```python
import numpy as np


def func(x):
    # заданная функция
    return 0.1 * x**2 - np.sin(x) + 0.1 * np.cos(x * 5) + 1.


def L(x):
    # квадратическая функция потерь
    return ((w[0] + w[1] * x + w[2] * x ** 2 + w[3] * x ** 3 + w[4] * x ** 4) - func(x)) ** 2

coord_x = np.arange(-5.0, 5.0, 0.1) # значения отсчетов по оси абсцисс
coord_y = func(coord_x) # значения функции по оси ординат

sz = len(coord_x) # общее число отсчетов

# здесь продолжайте программу
w = np.transpose(np.array([1.11, -0.26, 0.061, 0.0226, 0.00178]))
Q = 0
for i in range(sz):
    Q += L(coord_x[i])

Q *= 1 / sz
print(Q)

```
