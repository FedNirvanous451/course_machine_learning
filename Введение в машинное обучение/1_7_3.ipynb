```python
import numpy as np

x_test = np.array([(-5, 2), (-4, 6), (3, 2), (3, -3), (5, 5), (5, 2), (-1, 3)])
y_test = np.array([1, 1, 1, -1, -1, -1, -1])
w = np.array([-8/3, -2/3, 1])


#   <=> {1 ,  ;  0}
Q = sum([1 if (w[0] + w[1] * x_test[i][0] + w[2] * x_test[i][1]) * y_test[i] < 0 else 0 for i in range(len(x_test))])
print(Q)
```
