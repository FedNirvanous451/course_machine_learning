```python
import numpy as np

np.random.seed(0)
X = np.random.randint(0, 2, size=200)

def f_ones(j, i):
    # считает количество единиц до заданного индекса i
    return sum(x == 1 for x in X[j:i])

def gini(m, n):
    # считает индекс Джини для бинарного разбиения - передаем индексы разбиений
    all_ = abs(m - n)
    ones = f_ones(m, n)
    nulls = abs(m - n) - ones
    return 1 - (ones / all_) ** 2 - (nulls / all_) ** 2

S0 = gini(0, 200)
S1 = gini(0, 150)
S2 = gini(150, 200)

IG = S0 - 150 / 200 * S1 - 50 / 200 * S2
IG
```

```python
# Прикольное решение
import numpy as np

np.random.seed(0)
X = np.random.randint(0, 2, size=200)

# здесь продолжайте программу
def s(x):
    p1 = np.mean(x)
    p2 = np.mean(1 - x)
    
    S = 1 - (np.power(p1, 2) + np.power(p2, 2))
    return S

x1 = X[:150]
x2 = X[150:]

IG = s(X) - ( len(x1)/len(X) * s(x1) + len(x2)/len(X) * s(x2) )
IG
```
